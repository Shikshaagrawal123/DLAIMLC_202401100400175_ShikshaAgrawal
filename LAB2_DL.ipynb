```python

X = [75, 65, 80]
w = [4, 3, 5]
b = -1000
eta = 0.01
y = 1
```

```python
z = 0
for i in range(len(X)):
    z += w[i] * X[i]

z = z + b
print("Net input (z):", z)
```

```python
if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("Predicted Output:", y_pred)
```

```python
error = y - y_pred
print("Error:", error)
```

```python
if error != 0:
    for i in range(len(w)):
        w[i] = w[i] + (eta * error * X[i])
```

```python
b = b + (eta * error)
```

```python
z_new = 0
for i in range(len(X)):
    z_new += w[i] * X[i]

z_new = z_new + b

if z_new >= 0:
    y_pred_new = 1
else:
    y_pred_new = 0
```

```python
print("\nAfter Learning:")
print("Updated Weights:", w)
print("Updated Bias:", b)
print("New Net input (z):", z_new)
print("New Prediction:", y_pred_new)
```
